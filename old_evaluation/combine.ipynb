```python
import pandas as pd
```

```python
df2 = pd.read_csv('df2.csv')
df3 = pd.read_csv('df3.csv')
df4 = pd.read_csv('df4.csv')
df5 = pd.read_csv('df5.csv')
```

```python
df_combined = pd.concat([df2, df3, df4, df5], ignore_index=True)
df_combined.to_csv('results.csv')
df_combined.describe()
```

```python
for col, avg in df_combined.mean(numeric_only=True).items():
    print(f"Average of '{col}': {avg:.2f}")
```
